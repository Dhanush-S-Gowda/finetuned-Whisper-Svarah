```python
import os
import shutil
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict, Audio
from transformers import WhisperTokenizer, WhisperFeatureExtractor, WhisperProcessor, WhisperForConditionalGeneration
import torch
from torch.utils.data import DataLoader
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
import evaluate
```

```python
# Set paths
DATASET_PATH = "/content/svarah/svarah/audio"

# Create a list of all .wav and .txt files
wav_files = [f for f in os.listdir(DATASET_PATH) if f.endswith('.wav')]
txt_files = [f for f in os.listdir(DATASET_PATH) if f.endswith('.txt')]
```

```python
# Extract transcription content without file name and number
data = []
i = 0
for wav_file in wav_files:
    txt_file = wav_file.replace('.wav', '.txt')
    if txt_file in txt_files:
        with open(os.path.join(DATASET_PATH, txt_file), 'r') as f:
            transcription_line = f.read().strip()
            transcription_parts = transcription_line.split('\t')
            if len(transcription_parts) == 3:
                if i == 0:
                  print(transcription_parts[2])
                i+=1
                transcription = transcription_parts[2].strip()  # Take the third part (transcription)
            else:
                transcription = transcription_line  # Fallback in case of unexpected format
        data.append({"audio": os.path.join(DATASET_PATH, wav_file), "transcription": transcription})


# Split the data into train, test, and validation sets
train_data, test_val_data = train_test_split(data, test_size=0.2, random_state=42)
val_data, test_data = train_test_split(test_val_data, test_size=0.5, random_state=42)

# Reduce the size of the training data to a smaller subset (e.g., 20%)
reduced_train_data = train_data[:int(len(train_data) * 0.2)]

# Update the dataset with the reduced training data
dataset = DatasetDict({
    "train": Dataset.from_list(reduced_train_data),
    "validation": Dataset.from_list(val_data),
    "test": Dataset.from_list(test_data),
})


# Add audio features
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))
```

```python
model_name = "openai/whisper-small"
tokenizer = WhisperTokenizer.from_pretrained(model_name, language="en", task="transcribe")
feature_extractor = WhisperFeatureExtractor.from_pretrained(model_name)
processor = WhisperProcessor.from_pretrained(model_name, language="en", task="transcribe")
model = WhisperForConditionalGeneration.from_pretrained(model_name)
```

```python
# Preprocessing function
def preprocess_data(batch):
    audio = batch["audio"]["array"]
    # Process audio input
    inputs = processor.feature_extractor(audio, sampling_rate=16000, return_tensors="pt")
    # Process transcription
    labels = processor.tokenizer(batch["transcription"], return_tensors="pt", padding=True).input_ids
    # Update batch with processed inputs and labels
    batch["input_features"] = inputs.input_features.squeeze(0)
    batch["labels"] = labels.squeeze(0)
    return batch

# Apply preprocessing
dataset = dataset.map(preprocess_data, remove_columns=["audio", "transcription"], batched=False)

# Data collator for padding
class DataCollatorWhisper:
    def __call__(self, features):
        # Ensure input_features are tensors
        input_features = torch.stack([torch.tensor(f["input_features"]) for f in features])
        # Pad the labels appropriately
        labels = torch.nn.utils.rnn.pad_sequence(
            [torch.tensor(f["labels"]) for f in features], batch_first=True, padding_value=processor.tokenizer.pad_token_id
        )
        # Replace -100 (default padding) with pad_token_id for compatibility
        labels[labels == -100] = processor.tokenizer.pad_token_id
        return {"input_features": input_features, "labels": labels}


data_collator = DataCollatorWhisper()

# Training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=1e-5,
    per_device_train_batch_size=4,  # Reduced batch size
    per_device_eval_batch_size=4,  # Reduced evaluation batch size
    num_train_epochs=3,
    save_strategy="epoch",
    logging_dir="./logs",
    predict_with_generate=True,
    push_to_hub=False
)

```

```python
# Import required libraries
from evaluate import load
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

# Load WER metric
wer_metric = load("wer")

# Compute metrics function
def compute_metrics(pred):
    predictions = pred.predictions
    labels = pred.label_ids

    # Decode predictions and labels
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # WER calculation
    wer = wer_metric.compute(predictions=decoded_preds, references=decoded_labels)

    # Flatten predictions and labels for classification metrics
    true_labels = [label.strip() for label in decoded_labels]
    pred_labels = [pred.strip() for pred in decoded_preds]

    # Classification metrics
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, pred_labels, average="weighted")
    accuracy = accuracy_score(true_labels, pred_labels)

    # Return all metrics
    return {
        "wer": wer,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "accuracy": accuracy,
    }

# Trainer initialization
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    data_collator=data_collator,
    tokenizer=processor,
    compute_metrics=compute_metrics,
)

# Start training
trainer.train()

# Evaluate on the test set
test_results = trainer.evaluate(eval_dataset=dataset["test"])
print("Test Results:", test_results)
```

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor

# Load the fine-tuned model and processor
model = WhisperForConditionalGeneration.from_pretrained("/content/content/fine_tuned_whisper")
processor = WhisperProcessor.from_pretrained("/content/content/fine_tuned_whisper")
```

```python
import librosa
import torch

def predict_transcription(audio_path, model, processor):
    """
    Predict the transcript of an audio file using the fine-tuned Whisper model.

    Args:
        audio_path (str): Path to the audio file.
        model (WhisperForConditionalGeneration): Loaded Whisper model.
        processor (WhisperProcessor): Loaded Whisper processor.

    Returns:
        str: The predicted transcript.
    """
    # Load the audio file
    audio_array, sampling_rate = librosa.load(audio_path, sr=16000)

    # Preprocess the audio for Whisper
    inputs = processor(audio_array, sampling_rate=sampling_rate, return_tensors="pt")

    # Generate predictions
    with torch.no_grad():
        logits = model.generate(inputs["input_features"])

    # Decode the predictions into text
    transcription = processor.batch_decode(logits, skip_special_tokens=True)[0]
    return transcription
```

```python
# Path to your audio file
audio_file_path = "/content/Recording.m4a"

# Get the transcription
transcription = predict_transcription(audio_file_path, model, processor)

# Print the result
print("Predicted Transcript:", transcription)
```
